```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML
plt.rcParams['figure.figsize'] = [8, 6]
plt.rcParams['figure.dpi'] = 200
```

```python
data = np.genfromtxt("restart.HeisDM2D.out")

lattice = np.reshape(data[:,4:], (200, 200, 3))

plt.imshow(lattice[:,:,2], vmin=-1, vmax=1)
plt.colorbar()
```

```python
def calc_E(lattice):
    E = 0
    w = lattice.shape[0]
    h = lattice.shape[1]
    for x in range(w):
            for y in range(h):
                E -= lattice[x,y].dot(lattice[(x+1)%w,y])
                E -= lattice[x,y].dot(lattice[x,(y+1)%h])
                E -= np.cross(lattice[x,y], lattice[(x+1)%w,y]).dot([0, 1, 0])
                E -= np.cross(lattice[x,y], lattice[x,(y+1)%h]).dot([1, 0, 0])


    return E
```

```python
calc_E(lattice)
```
