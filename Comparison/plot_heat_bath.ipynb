```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
from pathlib import Path

from magn_config import *
```

```python
# lattice size
sizex, sizey = 200, 200

# physical parameters
Jexch = 1     # echange interaction
DMI   = 1.    # DMI interaction
Bext  = 0     # external magnetic field
Temp  = 1.0   # temperature
```

```python
num_config = 5  # number of configurations
```

```python
# magnetic configuration and attributes
df, MetaData = read_config("SKtest/hbf_D1_B0.h5")
sizex, sizey = MetaData["sizex"], MetaData["sizey"]
    
# calculate topological charge
Magnetizations = get_config_matrices(df, sizex, sizey)
for num, Mag in enumerate(Magnetizations, start=1):
    Q = get_topological_charge(Mag)
    print("Calculated topological charge: {:.3f}".format(Q))
         
    # plot magnetic configuration
    title = "HBF {:d}: Jexch = {Jexch:.1f}, D = {DMI:.2f}, B = {Bext:d}, T = {Temp:.1f}".format(num, **MetaData)
    plot_mz_map(Mag, pcol="black",
                xlim=(0., 200.), ylim=(0., 200.), title=title, 
                Positions=None, Triangulation=None,
                figname=None)
```
